# Which features have biggest betas?? 

```python
X = ordinals_train.drop(columns='saleprice')
y = ordinals_train['saleprice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.25,
                                                    random_state = 42)
```

```python
sc = StandardScaler()
```

```python
Z_train = sc.fit_transform(X_train, y_train)
```

```python
Z_test = sc.transform(X_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(Z_train, y_train)
```

```python
lr.score(Z_train, y_train)
```

```python
lr.score(Z_test, y_test)
```

```python
y_pred = lr.predict(Z_test)
y_pred
```

```python
rmse = mean_squared_error(y_test, y_pred)
rmse**.5
```

```python
lr.coef_
```
