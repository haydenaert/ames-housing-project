# Which unique values within features affects the mean of the sale price the most? 

```python
def get_unique_numbers(some_list): # source https://www.freecodecamp.org/news/python-unique-list-how-to-get-all-the-unique-values-in-a-list-or-array/
    '''This function takes in a list and returns a list of unique values'''
    list_of_unique_numbers = []

    unique_numbers = set(some_list)

    for number in unique_numbers:
        list_of_unique_numbers.append(number)

    return list_of_unique_numbers
```

```python
def important_str_cols(x, percent):
        '''This function receives a column with nominal values and returns the data that changes
    the mean of the sale price by more than 10%'''
        important_cols = []
        for i in nominal_train[x]:
            if abs(nominal_train['sale_price'].mean() - nominal_train[~nominal_train[x].str.contains(i)].sale_price.mean()) > abs(nominal_train['sale_price'].mean()*percent): 
                important_cols.append(i)
        return(get_unique_numbers(important_cols))
```

```python
# I ran this cell for each of my nominal columns 
# and I hand wrote down which unique values
# within each column affected the mean sale price 
# by more than 20% 
# For example, the unique value 'WD '
# in the 'sale_type' column affected 
# the mean of the sale price by at least 20%
important_str_cols('sale_type', .30)
```

# Use OneHotEncoder to turn nominal features in dummy variables 

```python
onehot = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
nominal_train = pd.DataFrame(onehot.fit_transform(nominal_train[['ms_subclass', 'ms_zoning', 'street', 'alley', 'land_contour',
       'lot_config', 'neighborhood', 'condition_1', 'condition_2', 'bldg_type',
       'house_style', 'roof_style', 'roof_matl', 'mas_vnr_type', 'foundation',
       'heating', 'central_air', 'garage_type', 'sale_type',]]).toarray(), 
                        columns=onehot.get_feature_names(['ms_subclass', 'ms_zoning', 'street', 'alley', 'land_contour',
       'lot_config', 'neighborhood', 'condition_1', 'condition_2', 'bldg_type',
       'house_style', 'roof_style', 'roof_matl', 'mas_vnr_type', 'foundation',
       'heating', 'central_air', 'garage_type', 'sale_type']))
```

```python
nominal_test = pd.DataFrame(onehot.fit_transform(nominal_test[['ms_subclass', 'ms_zoning', 'street', 'alley', 'land_contour',
       'lot_config', 'neighborhood', 'condition_1', 'condition_2', 'bldg_type',
       'house_style', 'roof_style', 'roof_matl', 'mas_vnr_type', 'foundation',
       'heating', 'central_air', 'garage_type', 'sale_type',]]).toarray(), 
                        columns=onehot.get_feature_names(['ms_subclass', 'ms_zoning', 'street', 'alley', 'land_contour',
       'lot_config', 'neighborhood', 'condition_1', 'condition_2', 'bldg_type',
       'house_style', 'roof_style', 'roof_matl', 'mas_vnr_type', 'foundation',
       'heating', 'central_air', 'garage_type', 'sale_type']))
```
